I have now imported the files and can load these into a pandas df

```python
import pandas as pd

df = pd.read_csv('EDA-DP.csv')
display(df.head(20))
```

There are three columns for salary, salary currency and salary in usd so I can remove the salary and salary currency columns

```python
df = df.drop(columns=['salary', 'salary_currency'], axis=1)
```

```python
df.head(20)
```

Now we have more data to look at we can compare salaries amongst those in higher positions and also the difference in how each job seems to pay

first, just to simplify, as all of the data we are looking at is for those working full time we can remove this column as well

```python
df = df.drop(columns=['employment_type'])
display(df.head())
```

```python
df.head(20)
```

Let's also remove the company location and employee_residence columns as they all say US

```python
df = df.drop(columns=['company_location', 'employee_residence'])
display(df.head())
```

```python
df = df.drop(columns=['work_year'])
display(df.head())
```

```python
df.head(20)
```

Now the columns are tidied up a bit we can delve deeper into differences between slaries

```python
average_salary_first_20 = df['salary_in_usd'].head(20).mean()
print(f"The average salary for the first 20 people is: ${average_salary_first_20:.2f}")
```

Now we have the average salary across these 20 individuals we can see who is making more and less than average and possibly why

```python
average_salary_ex = df[df['experience_level'] == 'EX']['salary_in_usd'].mean()
print(f"The average salary for individuals with 'EX' experience level is: ${average_salary_ex:.2f}")
```

```python
average_salary_se = df[df['experience_level'] == 'SE']['salary_in_usd'].mean()
print(f"The average salary for individuals with 'SE' experience level is: ${average_salary_se:.2f}")
```

```python
average_salary_mi = df[df['experience_level'] == 'MI']['salary_in_usd'].mean()
print(f"The average salary for individuals with 'MI' experience level is: ${average_salary_mi:.2f}")
```

```python
average_salary_en = df[df['experience_level'] == 'EN']['salary_in_usd'].mean()
print(f"The average salary for individuals with 'EN' experience level is: ${average_salary_en:.2f}")
```

As suspected, those with more experience get paid more than those with less experience, now we can visualise this with a graph

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a dictionary of experience levels and their average salaries
experience_levels = ['EX', 'SE', 'MI', 'EN']
average_salaries = [
    average_salary_ex,
    average_salary_se,
    average_salary_mi,
    average_salary_en
]

# Create a DataFrame for plotting
salary_data = pd.DataFrame({
    'Experience Level': experience_levels,
    'Average Salary (USD)': average_salaries
})

# Sort the data for better visualization, e.g., from lowest to highest salary
salary_data = salary_data.sort_values(by='Average Salary (USD)', ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(salary_data['Experience Level'], salary_data['Average Salary (USD)'], color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.title('Average Salary by Experience Level')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
