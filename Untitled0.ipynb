```python
import pandas as pd
df = pd.read_csv('loan_data.csv')
```

Now we have our df we can take a brief look at what we are working with. We will start by just looking at the initial 30 sets of data as there is too much data to evaluate at once

```python
df.head(30)
```

Now I have my dataframe I need to standardise my feature matrix and transform all numerical values. This is called z-score normalisation

```python
from sklearn.preprocessing import StandardScaler

# Assuming 'loan_status' is the target variable, select all other columns as features
X = df.drop('loan_status', axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Display the first few rows of the scaled features
print("Shape of X_scaled:", X_scaled.shape)
print("First 5 rows of X_scaled:")
print(X_scaled[:5])
```

```python
df.head(30)
```

Now we need to split the data. Using the code from week 4 in the github I will Train-Test split

```python
# split data into training and test
from sklearn.model_selection  import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, df['loan_status'], test_size = 0.2, random_state=1984, stratify=df['loan_status'])

# print the shapes to check everything is OK
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

This shows that 80% of the data is used for training and 20% is used for testing which we can see by the test_size=0.2

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Decision Tree Classifier with max_depth=5, min_samples_leaf=50, and min_samples_split=50
dtc = DecisionTreeClassifier(random_state=1984, max_depth=5, min_samples_leaf=50, min_samples_split=50)

# Train the model
dtc.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = dtc.predict(X_test)

# Evaluate the model
print("Decision Tree Classifier Performance (max_depth=5, min_samples_leaf=50, min_samples_split=50):")
print(f"Accuracy: {accuracy_score(Y_test, Y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred))
```

```python
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix
cm = confusion_matrix(Y_test, Y_pred)

print("Confusion Matrix for Decision Tree Model:")
print(cm)
```

I have made a decision tree which seems to have an accuracy of 0.84 which is pretty high but we want to test other models in order to compare and see which is the most accurate

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

# Assuming cm (confusion matrix) is already calculated from the previous step

# Normalize the confusion matrix over the true (actual) conditions
# This means each row sums to 1, showing the proportion of predictions for each actual class
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized, annot=True, fmt='.2%', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Normalized Confusion Matrix for Decision Tree Model')

plt.show()
```

We now have a confusion matrix for decision trees with it scaled down to percentages so that we are able to compare values more easily

Now lets look at the decision trees F1 model

```python
from sklearn.metrics import precision_recall_fscore_support

# Calculate precision, recall, and F1-score for Decision Tree
precision_dt, recall_dt, f1_score_dt, _ = precision_recall_fscore_support(Y_test, Y_pred, average='micro')

print("Decision Tree Micro-averaged Performance:")
print(f"Precision: {round(precision_dt, 4)}")
print(f"Recall: {round(recall_dt, 4)}")
print(f"F1-score: {round(f1_score_dt, 4)}")
print("\n")

precision_dt_macro, recall_dt_macro, f1_score_dt_macro, _ = precision_recall_fscore_support(Y_test, Y_pred, average='macro')

print("Decision Tree Macro-averaged Performance:")
print(f"Precision: {round(precision_dt_macro, 4)}")
print(f"Recall: {round(recall_dt_macro, 4)}")
print(f"F1-score: {round(f1_score_dt_macro, 4)}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Create a figure and axes for the plot
plt.figure(figsize=(20, 15)) # Adjust figure size for better readability

# Plot the decision tree
# feature_names are taken from the original DataFrame's columns before 'loan_status' was dropped
plot_tree(dtc, filled=True, feature_names=df.drop('loan_status', axis=1).columns.tolist(), class_names=['No Default', 'Default'], rounded=True, fontsize=10)
plt.title('Decision Tree Visualization', fontsize=16)
plt.show()
```

Let's look at logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Logistic Regression model
log_reg = LogisticRegression(random_state=1984, solver='liblinear') # 'liblinear' is good for small datasets

# Train the model
log_reg.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred_log_reg = log_reg.predict(X_test)

# Evaluate the model
print("Logistic Regression Model Performance:")
print(f"Accuracy: {accuracy_score(Y_test, Y_pred_log_reg):.4f}")
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred_log_reg))
```

A logistic regression model seems to have a similar accuracy to the decision tree - nlets evaluate its decision tree

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

# Calculate the confusion matrix for Logistic Regression
cm_log_reg = confusion_matrix(Y_test, Y_pred_log_reg)

# Normalize the confusion matrix over the true (actual) conditions
cm_normalized_log_reg = cm_log_reg.astype('float') / cm_log_reg.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized_log_reg, annot=True, fmt='.2%', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Normalized Confusion Matrix for Logistic Regression Model')

plt.show()
```

Whilst the F1 score for logistic regression is fairly similar to the decision tree model, the confusion matrix is split differently - with the decision tree making more accurate true responses than false and logistic regression making an average more correct predictions for true and false values
